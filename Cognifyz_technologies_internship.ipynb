# Task 2 : Descriptive analysis

```python
numeric_stats = df.describe()
print("Basic Statistical Measures for Numerical Columns:")
print(numeric_stats)
```

```python
# Explore the distribution of categorical variables
#Country Code
print("\nDistribution of 'Country Code' variable:")
country_distribution = df['Country Code'].value_counts()
print(country_distribution)
```

```python
#City
print("\nDistribution of 'City' variable:")
city_distribution = df['City'].value_counts()
print(city_distribution)
```

```python
#Cuisines
print("\nDistribution of 'Cuisines' variable:")
cuisine_distribution = df['Cuisines'].value_counts()
print(cuisine_distribution)
```

```python
# Identify the top cuisines and cities with the highest number of restaurants
top_cuisines = df['Cuisines'].value_counts().head(10)
print("\nTop 10 Cuisines:")
print(top_cuisines)

plt.figure(figsize=(12, 6))
top_cuisines.plot(kind='bar')
plt.title('Top cuisines with highest number of restaurants')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.show()
```

```python
top_cities = df['City'].value_counts().head(10)
print("\nTop 10 Cities with the Highest Number of Restaurants:")
print(top_cities)

plt.figure(figsize=(12, 6))
top_cities.plot(kind='bar', color='red')
plt.title('Top cities with highest number of restaurants')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.show()
```

# Task 3 : Geospatial analysis

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns


gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['Longitude'], df['Latitude']))

# Plot the locations of restaurants on map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(figsize=(15, 10), color='lightgray')
gdf.plot(ax=ax, marker='o', color='brown', markersize=10, label='Restaurants')
plt.title('Locations of Restaurants on the World Map')
plt.legend()
plt.show()


```

```python
# Analyze the distribution of restaurants across different countries
plt.figure(figsize=(12, 6))
sns.countplot(x='Country Code', data=df, order=df['Country Code'].value_counts().index)
plt.title('Distribution of Restaurants Across Countries')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

```python
# Determine if there is any correlation between the restaurant's location and its rating

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', data=df, palette='viridis', size='Aggregate rating', sizes=(20, 200))
plt.title('Correlation Between Restaurant Location and Rating')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Aggregate Rating')
plt.show()
```
